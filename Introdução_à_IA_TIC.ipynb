8. Gere uma nuvem de palavras para cada faixa etária de idade (['<20', '21–30', '31–40', '41–50', '51–60', '60+']) considerando o título das avaliações (campo 'review_title').

```python
# Calculando idade
df['age'] = 2025 - df['reviewer_birth_year']

# Classificando faixas etárias, recebe um número (idade) e devolve uma string com a faixa etária.
def faixa_etaria(idade):
    if idade < 20:
        return '<20'
    elif 21 <= idade <= 30:
        return '21-30'
    elif 31 <= idade <= 40:
        return '31-40'
    elif 41 <= idade <= 50:
        return '41-50'
    elif 51 <= idade <= 60:
        return '51-60'
    else:
        return '60+'

# aplica a função acima a cada valor de df['age']
# o apply chama faixa_etaria(idade) para cada linha e cria a nova coluna 'faixa_etaria'
df['faixa_etaria'] = df['age'].apply(faixa_etaria)

# Lista oredenada de faixas - garante ordem
faixas = ['<20', '21-30', '31-40', '41-50', '51-60', '60+']

for faixa in faixas:
    # Concatena todos os títulos dessa faixa etária em uma única string
    textos = " ".join(df[df['faixa_etaria'] == faixa]['review_title'].dropna().astype(str)) # 'dropna()' remove valores nulos e 'astype(str)' transforma tudo em string

    if textos.strip():  # só gera se tiver texto
        wc = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords_pt).generate(textos)

        # Plotar cada workcloud
        plt.imshow(wc, interpolation='bilinear')
        plt.axis('off')
        plt.title(f'Nuvem de Palavras - Faixa {faixa}')
        plt.show()
```

9. ***Desafio***: Gere um gráfico de barras com a distribuição de avaliações por faixa etária da idade. Considerar o seguinte range de faixas: '<20', '21–30', '31–40', '41–50', '51–60', '60+'

```python
# value_counts() conta quantas vezes cada faixa aparece no dataframe
# reindex(faixas, fill_value=0) reorganiza essas faixas de acordo com a ordem da lista 'faixas' (declarada no exercicio anterior) - força exibir na ordem que escolhi e não na ordem de contagem padrão.
# o fill_value faz com que se alguma faixa não aparecer no value_counts() (ex.: ninguém tem 60+ anos), ela aparece com valor 0 e não some.


df['faixa_etaria'].value_counts().reindex(faixas, fill_value=0).plot(kind='bar', color='pink')

plt.title('Distribuição de Avaliações por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Quantidade de Avaliações')
plt.show()
```

## Controlando a variabilidade da resposta com o parâmetro: Temperature

>[Configurações para a geração do texto](https://huggingface.co/docs/transformers/v4.52.3/en/main_classes/text_generation#transformers.GenerationConfig)

<img src=https://res.cloudinary.com/dwppkb069/image/upload/v1683736290/tips/images-03-temperature.mp4/03-temperature_30_03-25000-if-temperature-is-like-almost-0--were-going-to-have-a-very-sharp-peaked-distribution_wprxqo.png width="500">

Explicação dos próximos códigos

1️⃣ messages = [...]

* Aqui você cria uma lista com um dicionário dentro.
* Cada dicionário representa uma mensagem em um chat, como no estilo ChatGPT.

Campos do dicionário:

* "role" → indica quem está falando ("user", "assistant", "system", etc.)
* "content" → o texto da mensagem


2️⃣ pipe(messages)

* Pipeline de machine learning / NLP (ex.: Hugging Face Transformers).
* Ele recebe a lista de mensagens e processa para gerar uma resposta do modelo.
* O retorno geralmente é uma lista de resultados, mesmo que você mande uma mensagem só.

```python
messages = [
    {"role": "user", "content": "Tell me a good joke."}
]

# [0] pega o primeiro item da lista.
# "generated_text" normalmente é o campo que contém o texto que o modelo gerou.
# [1] indica que você está pegando o segundo elemento dentro de generated_text.
# ['content'] acessa o conteúdo textual real da resposta, que é o que o modelo escreveu como resposta da IA.

pipe(messages)[0]['generated_text'][1]['content']
```

3️⃣temperature controla aleatoriedade da resposta:

* 0 → respostas bem previsíveis, repetitivas
* 1 → respostas mais criativas e variadas
* 0.01 → praticamente sempre a mesma resposta

```python
messages = [
    {"role": "user", "content": "Me conte uma boa piada"},
]
pipe(messages, temperature=0.01)[0]["generated_text"][1]['content']
```
